```python
class Graph:
    def __init__(self, num_nodes):
        self.num_nodes = num_nodes
        self.adj_list = [[] for _ in range(num_nodes)]

    def add_edge(self, u, v):
        self.adj_list[u].append(v)
        # self.adj_list[v].append(u)

    def get_edges(self):
        edges = []
        for u in range(self.num_nodes):
            for v in self.adj_list[u]:
                edges.append((u, v))
        return edges

    def print_graph(self):
        for u in range(self.num_nodes):
            print(f"Node {u} is connected to: {self.adj_list[u]}")

# # Example usage
# if __name__ == "__main__":
#     num_nodes = 5
#     graph = Graph(num_nodes)

#     # Adding edges to the graph
#     graph.add_edge(3, 0)    
#     graph.add_edge(3, 1)
#     graph.add_edge(3, 2)
#     graph.add_edge(3, 3)
#     graph.add_edge(3, 4)
#     graph.add_edge(4, 0)
#     graph.add_edge(4, 1)
#     graph.add_edge(4, 2)
#     graph.add_edge(4, 3)
#     graph.add_edge(4, 4)

#     # Print the edges as a list of tuples
#     edges = graph.get_edges()
#     print("Edges of the graph:")
#     print(edges)

#     graph.print_graph()
```
